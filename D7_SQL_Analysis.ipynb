# SQL ANALYSIS

#### Total number of users who viewed, carted, and purchased

```python
%sql
SELECT event_type, COUNT(DISTINCT user_id) AS unique_users
FROM final_silver
GROUP BY event_type;
```

#### Conversion Rate

```python
%sql
SELECT
    (COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) * 1.0 /
     COUNT(DISTINCT CASE WHEN event_type = 'view' THEN user_id END)) AS conversion_rate
FROM final_silver;
```

#### Products with the highest views but lowest purchases

```python
%sql
SELECT 
    product_id,
    COUNT(CASE WHEN event_type = 'view' THEN 1 END) AS views,
    COUNT(CASE WHEN event_type = 'purchase' THEN 1 END) AS purchases
FROM final_silver
GROUP BY product_id
HAVING purchases = 0
ORDER BY views DESC;
```

#### Categorylevel conversion rate

```python
%sql
SELECT 
    category,
    COUNT(DISTINCT CASE WHEN event_type = 'view' THEN user_id END) AS viewers,
    COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) AS purchasers,
    (COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) * 1.0 /
     COUNT(DISTINCT CASE WHEN event_type = 'view' THEN user_id END)) AS category_conversion_rate
FROM final_silver
GROUP BY category
ORDER BY category_conversion_rate ASC;
```

#### Price impact on conversion

```python
%sql
SELECT 
    CASE 
        WHEN price < 50 THEN 'Low Price'
        WHEN price BETWEEN 50 AND 300 THEN 'Mid Price'
        ELSE 'High Price'
    END AS price_band,
    COUNT(DISTINCT CASE WHEN event_type = 'view' THEN user_id END) AS viewers,
    COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) AS purchasers,
    (COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) * 1.0 /
     COUNT(DISTINCT CASE WHEN event_type = 'view' THEN user_id END)) AS conversion_rate
FROM final_silver
GROUP BY price_band
ORDER BY conversion_rate ASC;
```

#### Brandlevel performance

```python
%sql
SELECT 
    brand,
    COUNT(CASE WHEN event_type = 'view' THEN 1 END) AS views,
    COUNT(CASE WHEN event_type = 'purchase' THEN 1 END) AS purchases,
    (COUNT(CASE WHEN event_type = 'purchase' THEN 1 END) * 1.0 /
     COUNT(CASE WHEN event_type = 'view' THEN 1 END)) AS brand_conversion_rate
FROM final_silver
GROUP BY brand
ORDER BY brand_conversion_rate ASC;
```
