# -	Governance: Unity Catalog setup with proper permissions

#### Load data

```python
df= spark.read.table("workspace.default.df_silver_events")
display(df.limit(2))
```

#### Constraint and Validation

```python
# Constraint and Validation
# Enforce positive price
df = df.filter(df["price"] > 0)

# Enforce valid event types
valid_events = ["view", "purchase"]
df = df.filter(df["event_type"].isin(valid_events))
```

#### Deduplication

```python
# Deduplication
df = df.dropDuplicates(["user_id", "product_id"])

df_silver_events=df
```

#### Access Control

```python
%sql
---Access control
GRANT SELECT ON TABLE df_silver_events TO `account users`;
```

#### Governance is now enforced with Unity Catalog, ensuring secure access, clear lineage, and proper permissions across my Gold tables.
